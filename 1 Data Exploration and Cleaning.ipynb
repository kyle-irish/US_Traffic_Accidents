top 7 values Clear, Overcast, Mostly Cloudy, Partly Cloudy, Scattered Clouds, Fair, Haze  would make sense to not be rainy conditions and make up over 90% of missing values. Will mark these as 0 rain.

```python
#Select the rows from traffic that are null for Precipitation(in) and Weather_Conditions are Clear, Overcast, Mostly Cloudy, Partly Cloudy, Scattered Clouds, Fair, Hazy

x = ['Clear','Overcast','Mostly Cloudy', 'Partly Cloudy', 'Scattered Clouds', 'Fair', 'Haze']

traffic[(traffic['Precipitation(in)'].isnull() == True) & (traffic['Weather_Condition'].isin(x))]
```

```python
#modify above selection to replace 'Precipitation(in)' value with 0

x = ['Clear','Overcast','Mostly Cloudy', 'Partly Cloudy', 'Scattered Clouds', 'Fair', 'Haze']
traffic.loc[(traffic['Precipitation(in)'].isnull() == True) & (traffic['Weather_Condition'].isin(x)),'Precipitation(in)'] = 0
```
